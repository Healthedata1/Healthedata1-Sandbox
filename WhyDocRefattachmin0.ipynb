# Why Document Reference.attachment.contentType Is Redundant And Thus Should be min=0

### In the majority of inline b64 cases, There is no need to declare the mime type separately 
when base64 encoding, it is already in the image data itself!

The example below shows how the jpg and pdf are encoded and unencoded without a separate declaration of the mime type.  This is out of the box stuff here, so there is not a need to reinvent the wheel

```python
from base64 import b64encode, b64decode

my_jpg = open('/Users/ehaas/Downloads/640px-Cat_August_2010-4.jpeg', 'rb')
my_b64_jpg = b64encode(my_jpg.read())
my_jpg.close()

print(f'{my_b64_jpg[:100]}...')
```

```python
my_pdf = open('/Users/ehaas/Downloads/640px-Cat_August_2010-4.pdf', 'rb')
my_b64_pdf = b64encode(my_pdf.read())
my_pdf.close()

print(f'{my_b64_pdf[:100]}...')
```

```python
from IPython.display import Image
cat_pix= b64decode(my_b64_jpg)
Image(cat_pix)
```

```python
my_cat_pix = open('/Users/ehaas/Downloads/my_cat_pix.pdf', 'wb')
my_cat_pix.write(b64decode(my_b64_pdf))
my_cat_pix.close()

print('''Sadly PDFs are not supported out-of-the-box in Jupyter for many browsers (e.g. Brave, Chrome, Safari...
... So I saved it and opened it locally in a pdf view for this screen shot
''')

my_cat_pdf = open('/Users/ehaas/Downloads/my_cat_pdf.png', 'rb')
cat_pix=my_cat_pdf.read()
my_cat_pdf.close()

Image(cat_pix)
```
